# Separating the data for analysis

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

# Statistical Measure of the Data

```python
df.groupby('Class').mean()
```

# Under-Sampling

Under Sampling-Building a dataset which has equal number of Legit & Fraud Data

```python
legit_sample=legit.sample(n=492)
```

# Concat Two Dataframes

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

# Splitting the data into X & Y

```python
x=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
```

```python
print(x)
```

```python
print(y)
```

# Splitting the data into training & testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

# Training model through Logistic Regression

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

# Accuracy Score

**Accuracy of Training Data**

```python
x_train_pred=lr.predict(x_train)
x_train_accuracy=accuracy_score(x_train_pred,y_train)
print('Accuracy of Training Data:',x_train_accuracy)
```

**Accuracy of Testing Data**

```python
x_test_pred=lr.predict(x_test)
x_test_accuracy=accuracy_score(x_test_pred,y_test)
print('Accuracy of Testing Data:',x_test_accuracy)
```
